## Be careful when copying arrays 

```python
a = np.array([1,2,3])
b = a 
print(a, b)

b[0] = 100
print(a, b)

# They are pointing the same thing
# To avoid this, use copy() method

c = np.array([1,2,3])
d = c.copy()
print(c, d)

d[0] = 100
print(c, d)
```
